### **1**.	Vamos acceder a una base de datos del repositorio UCI Machine Learning Repository. La información se encuentra disponible en https://archive.ics.uci.edu/ml/datasets/Car+Evaluation. Empezaremos importando la librería ‘pandas’.

```python
import pandas as pd
```

### 2.	Primero extraemos la información del enlace anterior pudiendo almacenarlo en una variable (‘url’). Debemos recordar que este enlace debe estar concatenado con el nombre de la basime de datos (‘car.data’). Luego, lo leemos con el método ‘read_csv( )’.

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/car/car.data'
```

```python
data = pd.read_csv(url)
data
```

### 3.	Si en caso, los nombres de las variables no sean los que se muestran en el enlace o simplemente se le quiere asignar otros nombres, podemos modificar el parámetro ‘names’ a manera de una lista que contenga los nombres de los atributos.

```python
data = pd.read_csv(url, names=['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety', 'acceptability'])
data
```

### 4.	Asimismo, es posible extraer los nombres de los atributos de la base de datos original por medio de cualquiera de los métodos mostrados.

```python
from typing import List
#Extraer columnas
list(data)
list(data.columns)
data.columns.tolist()
list(data.columns.values)
```

### 5.	Para medir el rendimiento de los métodos anteriores en términos de tiempo podemos usar la función ‘timeit’ de la librería ‘timeit’.

```python
#Medicion de rendimiento
from timeit import timeit
t1 = timeit(lambda: list(data))
t2 = timeit(lambda: list(data.columns))
t3 = timeit(lambda: data.columns.tolist())
t4 = timeit(lambda: list(data.columns.values))
print(t1,t2,t3,t4)
```

### 6.	Usando el método ‘dtype’ para cada atributo, podemos ver su naturaleza. Si quisiéramos ver esto para todos los atributos, usaremos el método ‘dtypes’.

```python
data['buying'].dtype
data['persons'].dtype
data.dtypes
```


### 7.	Podemos observar la cantidad de categorías de un atributo mediante el método ‘unique( )’.


```python
data["doors"].unique()
```

### 8.	Si en el paso anterior, la naturaleza de los atributos no coincide en cómo se está definiendo en el repositorio web, podemos cambiar esto por medio del método ‘astype( )’ aplicado sobre el atributo junto con su tipo.

```python
#Cambiando tipo de datos
data['doors'] = data['doors'].astype('category')
data['persons'] = data['persons'].astype('category')
data.dtypes
```

### 9.	Para cambiar el nombre a las categorías debemos importar la librería ‘numpy’ que trabaja con arreglos y luego utilizar el método ‘where( )’ indicando al atributo de la data, el nuevo nombre para cada categoría.

```python
import numpy as np
data["lug_boot"] = np.where(data["lug_boot"] =="small", "pequeño", data["lug_boot"])
data["lug_boot"] = np.where(data["lug_boot"] =="med", "mediano", data["lug_boot"])
data["lug_boot"] = np.where(data["lug_boot"] =="big", "grande", data["lug_boot"])
data["lug_boot"].unique()
```

### 10.	Con el método ‘.value_count( )’ podemos contabilizar las frecuencias de cada categoría.

```python
data['doors'].value_counts(dropna=False)
```

### 11.	También podemos juntar dos o más categorías mediante el método ‘where’ escribiendo la misma categoría para las que son diferentes.

```python
data["doors"] = np.where(data["doors"] == "2", "3 a menos", data["doors"])
data["doors"] = np.where(data["doors"] == "3", "3 a menos", data["doors"])

data["doors"] = np.where(data["doors"] == "4", "4 a más", data["doors"])
data["doors"] = np.where(data["doors"] == "5more", "4 a más", data["doors"])
data["doors"].unique()
```

### 12.	Otra forma de cambiar los nombres de los encabezados es mediante el método ‘rename’ escribiendo en forma de diccionario, el nombre de la variable a cambiar (‘lug_boot’) y luego el nuevo nombre (‘trunk’). Además, debemos asignar el valor ‘axis=1’ y si queremos cambiar nombres a las filas usamos ‘axis=0’.

```python
data.rename({'lug_boot': 'trunk'}, axis=1, inplace=True)
data
```
